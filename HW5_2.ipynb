```python
import pandas as pd

# Read in Dataset
browsers = pd.read_csv(
    "datasets/worldwide_browser_trends.csv",
    parse_dates=["Month"],
    index_col="Month"
)

# Test to aalyze data
browsers.info()
browsers.head()
```

```python
#Initial Visualization 
browsers.plot()
```

```python
###1

# Calculate six-month rolling average for each browser
rolling_six = browsers.rolling(window=6).mean()
rolling_six.head(10)
rolling_six.plot(title="6-Month Rolling Average of Browser Search Interest")
```

```python
###2

# Calculate month-over-month percentage change with 3-month lag (i.e., quarterly change)
pct_change_quarterly = browsers.pct_change(periods=3) * 100

# Filter to include only dates from 2009 onward
pct_change_quarterly = pct_change_quarterly.loc['2009':]


pct_change_quarterly.plot(subplots=True, figsize=(12,8))
```

```python
###3

# Create a DataFrame to hold Chrome search interest for each target year
chrome_trends = pd.DataFrame()

# Define the years we want to extract
years = [2009, 2012, 2015, 2018]

# Loop through each year, extract monthly Google Chrome data, and assign to new column
for year in years:
    chrome_trends[str(year)] = browsers.loc[str(year), 'Google Chrome'].reset_index(drop=True)

# Plot the results
chrome_trends.plot(title="Chrome Search Performance in 2009, 2012, 2015 & 2018")


```
